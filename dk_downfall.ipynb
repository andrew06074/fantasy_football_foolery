```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('dk_downfall.csv')
```

```python
df
```

```python
#my projected points compared to actual points, sanity shown for medical purposes
plt.plot(df.Week, df.Projected)
plt.plot(df.Week, df.Actual )
plt.plot([1,8,11],[150,100,0])
plt.legend(['Andrew\'s projected points','Andrew\'s actual points','Andrew\'s sanity'])
plt.xticks(df.Week)
plt.ylabel('Points')
plt.xlabel('Week')
plt.show()
```

```python
#my points compared to my opponents points
plt.plot(df.Vs, df.Actual)
plt.plot(df.Vs, df.Opp)
plt.legend(['Andrew\'s points','Opponent\'s points'])
plt.xticks(rotation = 90)
plt.ylabel('Point')
plt.show()
```

```python
#point differential in win/loss
y = df.Actual - df.Opp
plt.plot(df.Vs, y)
plt.plot([0,10],[0,0])
plt.legend(['Points differential','Win(above)/Loss(below)'])
plt.xticks(rotation = 90)
for a,b in zip(df.Vs, y.round(2)):
    plt.text(a,b, str(b))
plt.ylabel('Point differential')
plt.show()
```

```python
#highlight weeks aftrer dropping MG3
plt.plot(df.Week[7:11], df.Actual[7:11] - df.Opp[7:11])
plt.plot([8,11],[0,0])
plt.legend(['MG3 curse','Loss line'])
plt.xticks(df.Week[7:11])
plt.ylabel('Point differential')
plt.xlabel('Week')
plt.show()
```

```python
#highlight the effects of the MG3 curse
plt.plot(df.Week[7:11], df.Actual[7:11] - df.Opp[7:11])
plt.plot([8,11],[0,0])
plt.plot(df.Week[7:11],df.MG3[7:11])
plt.legend(['MG3 curse','Loss line','MG3 points'])
plt.xticks(df.Week[7:11])
plt.xlabel('Week')
plt.ylabel('Point differential / MG3 poi')
plt.show()
```
