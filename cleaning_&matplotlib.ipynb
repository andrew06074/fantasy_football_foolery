```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
proj_data = pd.read_csv('draft_day_proj.csv')
proj_data.columns = ['Rank', 'Team', 'Record', 'Pct', 'Pts', 'Pts_Against']
```

```python
current = pd.read_csv('current_standings.csv')
current.columns = ['Rank', 'Team', 'Record', 'Pts_For', 'Pts_Agnst', 'Streak', 'Waiver',
       'Moves']
```

```python
proj_data = proj_data.sort_values('Team')
current = current.sort_values('Team')
current.reset_index(drop=True, inplace= True)
proj_data.reset_index(drop=True, inplace=True)
```

```python
x = ['A Team Has No Name','Brady Respecter','DK Bestcalf','Honk','John\'s Team','Mike\'s Team','Mousecop','OhKaeptainMyKaeptain','Rick' ,'Sea Org','Smitty W','The_Donta']
plt.plot(x, proj_data.Pts * .7692307692307692)
plt.plot(x, current.Pts_For)
plt.legend(['Projected', 'Actual'])
plt.xticks(rotation=90)
plt.tick_params(labelsize=16)
plt.ylabel('Total team points')
plt.title('Projected points vs Actual points (through week 10)')
plt.show()
```
