### Imports and read file

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 7, 20
df = pd.read_csv('2020_draft.csv')
```

### Clean Rankings

```python
df['Position'] = df['Pos_Rank']
df['Position'] = df['Position'].str[0:2]
```

### Create sub df of running backs and export to csv

```python
df_rb = df.loc[df['Position'] == 'RB']
df_rb = df_rb.drop(columns=['Pos_Rank','Depth','Auction'])
df_rb = df_rb.sort_values('2020_FP',ascending = False)
df_rb.reset_index(drop=True, inplace= True)
df_rb.to_csv('2020_RB.csv', index = False)
```

# The following is a breakdown based on player position. I will compare the players projected points for the 2020 season to how they preformed in 2019.
### My goal is to use the created visulizations to assist with my draft

# Top RB

```python
df_rb = df_rb.head(35).iloc[::-1]

x = df_rb['Player']

proj2020 = df_rb['2020_FP']
acc2019 = df_rb['2019_FP']


ind = np.arange(len(x))
width = 0.35

fig, ax = plt.subplots()

plt.barh(ind, proj2020, width, color='blue', label='2020 Projected')
plt.barh(ind + width, acc2019, width, color='gray', label='2019 Actual')


ax.set(yticks=ind+width, yticklabels=x)

plt.legend(loc='best')

plt.show()
```
